```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import pytorch_lightning as pl
from torchvision import datasets,transforms
from torch.utils.data import random_split, DataLoader
```

```python
class NN(pl.LightningModule):
  def __init__(self):
    super(NN,self).__init__()
    self.fc1 = nn.Linear(28*28,256)
    self.fc2 = nn.Linear(256,128)
    self.fc3 = nn.Linear(128,10)
    self.Loss = nn.CrossEntropyLoss()

  def forward(self,x):
    x = torch.flatten(x,1)
    #x = x.view(-1,28*28)
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    return F.sigmoid(self.fc3(x))

  def configure_optimizers(self):
    return optim.Adam(self.parameters(),lr=0.001)

  def training_step(self,tbatch,tidx):
    x,y = tbatch
    output = self(x)
    loss = self.Loss(output,y)
    return loss

  def validation_step(self, vbatch,vidx):
    x,y = vbatch
    output = self(x)
    loss = self.Loss(output,y)
```

```python
class Data(pl.LightningDataModule):
    def prepare_data(self):
        transform=transforms.Compose([
            transforms.ToTensor()
        ])
      
        self.train_data = datasets.MNIST('', train=True, download=True, transform=transform)
        self.test_data = datasets.MNIST('', train=False, download=True, transform=transform)

    def train_dataloader(self):
        return DataLoader(self.train_data, batch_size= 32, shuffle=True)

    def val_dataloader(self):
        return DataLoader(self.test_data, batch_size= 32, shuffle=False)
```

```python
model = NN()
trainer = pl.Trainer(max_epochs=5)
data = Data()
trainer.fit(model,data)
```
